```python
x=df_train_transformed_4.drop('Item_Outlet_Sales',axis=1)
```

```python
y = df_train_transformed_4['Item_Outlet_Sales']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.33, random_state=10)
```

```python
x_train
```

```python
y_test
```

```python
#Feature SCaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm=MinMaxScaler().fit(x_train)
x_train_norm=norm.transform(x_train)
```

```python
x_train.head(10)
```

```python
scaled_features_df=pd.DataFrame(x_train_norm,index=x_train.index ,columns=x_train.columns)
```

```python
scaled_features_df.head(10)
```

```python
#all features are in the same range
```

```python
x_train=scaled_features_df
```

```python
x_train
```

```python
norm= MinMaxScaler().fit(x_test)
x_test_norm=norm.transform(x_test)
```

```python
x_test_norm
```

```python
scaled_features_df_1=pd.DataFrame(x_test_norm,index=x_test.index ,columns=x_test.columns)
```

```python
scaled_features_df_1.head(10)
```

```python
x_test=scaled_features_df_1
```

```python
# we did feature selection , feature encoding , feature formatting,feature splitting, feature scalling . now we have to make the model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
import seaborn as sns
sns.distplot((y_test-y_pred))
#the below distribution plot depicts the variation in the data distribution
```

```python
print("training acc:",reg.score(x_train,y_train))
print("testing acc:",reg.score(x_test,y_test))
```

```python
#Nearly identical performance on the training set and test set is a good outcome, it means the model is doing what it's supposed to do
```
